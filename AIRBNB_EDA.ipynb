```python
numcols = [col for col in df.columns if df[col].dtype != 'O']

def identify_outliers(df, numcols):
    outlier_indices = []
    for col in numcols:
        quantile_25 = df[col].quantile(0.25)
        quantile_75 = df[col].quantile(0.75)
        iqr = quantile_75 - quantile_25


        lower_bound = quantile_25 - 1.5 * iqr
        upper_bound = quantile_75 + 1.5 * iqr


        outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)].index
        outlier_indices.extend(outliers)


    outlier_indices = list(set(outlier_indices))


    is_outlier = df.index.isin(outlier_indices)

    return is_outlier


outlier_mask = identify_outliers(df, numcols)
df_no_outliers = df[~outlier_mask]

```
